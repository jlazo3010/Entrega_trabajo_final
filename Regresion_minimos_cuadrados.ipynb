```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.linear_model import LinearRegression
```

```python
indicador_empleo = pd.read_csv('/content/drive/MyDrive/Analisis_numerico/Empleo_porcentaje.csv')
indicador_inflacion = pd.read_csv('/content/drive/MyDrive/Analisis_numerico/Inflacion_porcentaje.csv')
indicador_produccion = pd.read_csv('/content/drive/MyDrive/Analisis_numerico/ProduccionConstruccion.csv')
```

```python
#Limpieza de Na´s
indicador_empleo = indicador_empleo.dropna(axis=0, how='any')
indicador_inflacion = indicador_inflacion.dropna(axis=0, how='any')
indicador_produccion = indicador_produccion.dropna(axis=0, how='any')
```

```python
def regresion_lineal(df):
  x = np.arange(0,len(df.iloc[:,1])).reshape(-1,1)
  y = df.iloc[:,1] 
  regresion_lineal = LinearRegression() #generamos el modelo
  regresion_lineal.fit(x, y) #entrenamos el modelo
  Y_predict = regresion_lineal.predict(x) 
  plt.figure(figsize=(10,7))
  plt.scatter(x, y)
  plt.plot(x, Y_predict, color='red', linewidth =3)
  plt.title('Regresión Lineal Simple')
```

```python
regresion_lineal(indicador_empleo) #Regresión por minimos cuadrados de indicadores de empleo
```

```python
regresion_lineal(indicador_inflacion) #Regresión por minimos cuadrados de indicadores de inflación
```

```python
regresion_lineal(indicador_produccion) #Regresión por minimos cuadrados de indicadores de producción
```

```python
def regresion_lineal_prediccion(df):
  x = np.arange(0,len(df.iloc[:,1])).reshape(-1,1)
  y = df.iloc[:,1] 
  x_pred = np.arange(0, len(df.iloc[:,1])+10).reshape(-1,1) #Estos datos predeciremos
  regresion_lineal = LinearRegression() #generamos el modelo
  regresion_lineal.fit(x, y) #entrenamos el modelo
  Y_predict = regresion_lineal.predict(x_pred) 
  plt.figure(figsize=(10,7))
  plt.scatter(x, y)
  plt.plot(x_pred, Y_predict, color='red', linewidth =3)
  plt.title('Regresión Lineal Simple')
```

```python
regresion_lineal_prediccion(indicador_empleo) #Regresión por minimos cuadrados de indicadores de empleo con predicción
```

```python
regresion_lineal_prediccion(indicador_inflacion) #Regresión por minimos cuadrados de indicadores de inflación con predicción
```

```python
regresion_lineal_prediccion(indicador_produccion) #Regresión por minimos cuadrados de indicadores de producción con predicción
```
